```python
import cv2
```

```python
cls=cv2.CascadeClassifier("C:/Users/CG-DTE/AppData/Local/Programs/Python/Python37/Scripts/OpenCV/counting bikes/two_wheeler.xml")
cap=cv2.VideoCapture("bikes.mp4")
c=0
while True:
    ret,frame=cap.read()
    blur=cv2.GaussianBlur(frame,(3,3),0)
    grey=cv2.cvtColor(blur,cv2.COLOR_BGR2GRAY)
    face=cls.detectMultiScale(grey,1.1,1)
    H,W=frame.shape[0:2]
#     print(H,W)
    cv2.putText(frame,"Count:",(100,200),cv2.FONT_HERSHEY_PLAIN,2,(255,255,0),3)
    cv2.line(frame,(0,H-200),(W,H-200),(0,0,255),2)

    for (x,y,w,h) in face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),1)
        cv2.putText(frame,"Bike",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.5,(255,255,190),2)
        print(x,y)
        if  (y+h/2)<((H-200)+6) and (y+h/2)>((H-200)-6):
            c+=1
        cv2.putText(frame,str(c),(100+120,200),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),2)
            
        
    cv2.imshow("img",frame)
    if cv2.waitKey(1)==13:
        break
cap.release()
cv2.destroyAllWindows()
```
